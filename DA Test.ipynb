## Khách hàng được khảo sát nhiều nhất ở tầm tuổi từ 20 đến 60 tuổi

```python
categorical_features=[feature for feature in df.columns if ((df[feature].dtypes=='O') & (feature not in ['y']))]
categorical_features
```

```python
plt.figure(figsize=(15,80))
plotnumber =1
for categorical_feature in categorical_features:
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(y=categorical_feature,data=df)
    plt.xlabel(categorical_feature)
    plt.title(categorical_feature)
    plotnumber+=1
plt.show()
```

## Nhận xét về số đối tượng đã ghi nhận
- Về job, admin. là công việc được ghi nhận nhiều nhất, ít nhất là đối tượng unknown, sau đó là học sinh
- Về marital, khách hàng đã kết hôn chiếm phần nhiều nhất, ít nhất là unknown, ít thứ 2 là đã ly dị
- Về trình độ học vấn education, ghi nhận nhiều nhất là những khách hàng có bằng đại học, ít nhất là các khách hàng không biết chữ.
- Hầu hết các giá trị đều được ghi nhận tại defaut = yes
- Dữ liệu của tháng 5 là nhiều nhất, ít nhất là tháng 12.
- Về poutcomes, giá trị không tồn tại chiếm nhiều nhất, đa phần dữ liệu lấy được đều từ những khách hàng lần đầu tham gia chiến dịch marketting.

```python
for categorical_feature in categorical_features:
    sns.catplot(x='y', col=categorical_feature, kind='count', data= df)
plt.show()
```

```python
for categorical_feature in categorical_features:
    print(df.groupby(['y',categorical_feature]).size())
```

```python
joby = pd.crosstab(df['job'], df['y'])
joby.div(joby.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, figsize=(15,7))
plt.xlabel('job')
plt.ylabel('percentage')
```

```python
monthy = pd.crosstab(df['month'], df['y'])
monthy.div(monthy.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, figsize=(15,7))
plt.xlabel('month')
plt.ylabel('percentage')
```

```python
defaulty = pd.crosstab(df['default'], df['y'])
defaulty.div(defaulty.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, figsize=(15,7))
plt.xlabel('default')
plt.ylabel('percentage')
```

## Nhận xét về tỉ lệ khách hàng có đăng kí gói term deposit dựa vào các thuộc tính
- Job: khách hàng thuộc nhóm retired và student có tỉ lệ đồng ý đăng kí gói cao
- housing và loan: khách hàng có khoản vay trả tiền nhà và các khoản vay cá nhân có xu hướng ít đồng ý với gói
- poutcome: nếu kết quả của chiến dịch trước là thành công thì khách hàng có tỉ lệ cao hơn sẽ đăng kí gói.
- Tháng 3, 9, 10, 12 khách hàng có tỉ lệ đăng kí gói cao.
- Tháng 5 có nhiều record được ghi lại nhưng tỉ lệ đăng kí gói thấp.
- Khách hàng không có defaut từ trước (khoản không thanh toán được) sẽ có xu hướng đăng kí gói nhiều hơn, trong khi nếu đã có khoản không thanh toán được trong quá khứ thì không có ai trong khảo sát đăng kí gói.
